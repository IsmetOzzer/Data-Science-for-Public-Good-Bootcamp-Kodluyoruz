```python
#Best wine, mmm 
filt = (wine["quality"] >= 8)
```

```python
best_wine = wine.loc[filt]
```

```python
best_wine 
```

```python
# Best wines with least amount of acid 
wine.loc[(wine["citric acid"] <= .3) & (wine["quality"] >= 8)]
```

```python
plt.plot(best_wine["fixed acidity"],
         best_wine["pH"], 
         "o",
        color = "red")
plt.title("pH increases as fixed acidity decreases")
plt.xlabel("pH")
plt.ylabel("Fixed acidity")
```

```python
#Capitalize the column names in best_wine
best_wine.columns = [col.capitalize() for col in best_wine.columns]
```

```python
#change the quality of the 4th index to 6
wine.loc[4, "quality"] = 6

wine.head()
```

```python
#Create a new column called rating and assign "good" if the quality is > 6, else assign "bad"

wine["rating"] = np.where(wine["quality"] > 6, "good","bad")
```

```python
wine.head(10)
```

```python
plt.bar(wine["rating"],
        wine["alcohol"])
plt.title("Rating by alcohol levels")
plt.xlabel("Rating")
plt.ylabel("Alcohol levels")
```
